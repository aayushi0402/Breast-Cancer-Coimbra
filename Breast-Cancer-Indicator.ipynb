# Breast Cancer Indicatior 

The dataset has been obtained from UCI Machine Learning repository.
Clinical features were observed or measured for 64 patients with breast cancer and 52 healthy controls.
There are 10 predictors, all quantitative, and a binary dependent variable, indicating the presence or absence of breast cancer. 

The predictors are anthropometric data and parameters which can be gathered in routine blood analysis. 
Prediction models based on these predictors, if accurate, can potentially be used as a biomarker of breast cancer.



Quantitative Attributes: 
Age (years), 
BMI (kg/m2), 
Glucose (mg/dL) ,
Insulin (µU/mL), 
HOMA ,
Leptin (ng/mL) ,
Adiponectin (µg/mL) ,
Resistin (ng/mL), 
MCP-1(pg/dL) ,

Labels: 
1=Healthy controls, 
2=Patients


Reference Link:
https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Coimbra

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('dataR2.csv')
data
```

### Getting to know the dataset

```python
data.info()
```

### Number of  Healthy Controls and Patients in the sample

```python
ax = sns.countplot(x = "Classification", data= data,edgecolor=sns.color_palette("bright", 2),facecolor = (0,0,0,0), linewidth = 10)
```

```python
data.loc[data['Classification'] == 1] .describe()
```

```python
data.loc[data['Classification'] == 2] .describe()
```

We see that the mean of Glucose, Insulin,HOMA, Resistin and MCP1 is greater in Cancer Patients than in Healthy Controls.
The standard deviation of Glucose and MCP1 is also very high in Cancer Patients when compared to Healthy Controls.

### Age Distribution of Sample

```python
ax = data.Age.hist(bins=20,color='teal',alpha=0.8)
ax.set(xlabel='Age',ylabel='Count')
plt.show()
```

### Age vs Classification Distribution

```python
sns.kdeplot(data.Age[data.Classification == 1],color="purple",shade=True)
sns.kdeplot(data.Age[data.Classification == 2],color="blue",shade=True)
plt.legend(['Healthy Controls', 'Patients'])
plt.title('Age vs Classification')
plt.show()
```

### Age and Body Mass Index[BMI] Distribution

```python
sns.kdeplot(data['Age'],data['BMI'],shade = True, cmap = 'Greens')
```

### Pairwise Plot

```python
sns.pairplot(data)
```
